```python
alpha = 0.05  # significance level
for col in reg_df.columns:
    stat, p = stats.shapiro(reg_df[col])
    if p < alpha:
        print(f"{col} is not normally distributed (p={p:.4f})")
    else:
        print(f"{col} is normally distributed (p={p:.4f})")
```

```python
def table(df, endogs, exogs, digits):
    df_ = pd.DataFrame(index = ["const"] + exogs + ["N","R2","F"], columns = endogs)
    for endog in endogs:
        sub_df = df[exogs + [endog]].dropna()
        X = sub_df[exogs]
        X = sm.add_constant(X)
        y = sub_df[endog]
        model = sm.OLS(y, X)
        results = model.fit(cov_type = 'HC3')
        results_df = pd.DataFrame(results.params) 
        results_df["Pvalues"] = results.pvalues
        results_df["std error"] = results.HC3_se
        results_df["Stars"] = results_df["Pvalues"].apply(lambda x: "***" if x < 0.01 else ("**" if x < 0.05 else ("*" if x < 0.1 else ""))) 
        results_df["Final"] = results_df[0].round(digits).astype(str) + results_df["Stars"] + " [" + results_df["std error"].round(digits).astype(str) + "]"
        f_val = str(round(results.fvalue,3))
        p = results.f_pvalue
        if p < 0.01:
            f_val += "***"
        elif p < 0.05:
            f_val += "**"
        elif p < 0.1:
            f_val += "*"
        df_[endog] = results_df["Final"].tolist() + [len(sub_df), round(results.rsquared,3), f_val]
    return df_
```

```python
reg_df["Loans per Emp Bus"] = reg_df["Loans per Emp Bus_1"] + reg_df["Loans per Emp Bus_2"] + reg_df["Loans per Emp Bus_3"]
reg_df["Avg Loan Amt per Emp"] = reg_df['Avg Loan Amt per Emp_1'] + reg_df['Avg Loan Amt per Emp_2'] + reg_df['Avg Loan Amt per Emp_3']
```
