```python
import geopandas as gpd
import numpy as np
import pandas as pd
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.animation import FuncAnimation
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.ticker as mtick
```

```python
# Load the first shapefile
gdf1 = gpd.read_file("cb_2020_us_aiannh_500k.shp")

# Load the second shapefile
gdf2 = gpd.read_file("cb_2020_us_zcta520_500k.shp")

# Load the third shapefile
gdf3 = gpd.read_file("cb_2020_us_nation_5m.shp")

# Create a plot
fig, ax = plt.subplots(figsize=(100, 100))

# Plot the third GeoDataFrame
gdf3.plot(ax=ax, color='white', edgecolor='k', alpha = 0.15)

# Plot the first GeoDataFrame
gdf1.plot(ax=ax, color='blue', edgecolor='blue', alpha = 0.3)

# Plot the second GeoDataFrame
gdf2.plot(ax=ax, color='red', edgecolor='red', alpha = 0.15)

# Show the plot with a legend
plt.legend()
plt.show()
```
