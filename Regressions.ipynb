```python
def format_float(val, digits):
    format_str = f'{{:.{digits}f}}'
    return format_str.format(val)

def table(df, endogs, exogs, digits):
    # Set display format to avoid scientific notation for DataFrame display
    pd.options.display.float_format = lambda x: format_float(x, digits)

    df_ = pd.DataFrame(index=["const"] + exogs + ["Mean", "N", "R2", "F", 'AIC', 'BIC'], columns=endogs)
    for endog in endogs:
        sub_df = df[exogs + [endog, "Rural_i"]].dropna()
        print(sub_df["Min Share Excl B"].mean())
        print(sub_df["Black Share"].mean())
        print(sub_df[sub_df["Rural_i"] == 0]["Min Share Excl B"].mean())
        print(sub_df[sub_df["Rural_i"] == 1]["Min Share Excl B"].mean())
        X = sub_df[exogs]
        X = sm.add_constant(X)
        y = sub_df[endog]
        model = sm.OLS(y, X)
        results = model.fit(cov_type='HC3')
        results_df = pd.DataFrame(results.params)
        results_df["Pvalues"] = results.pvalues
        results_df["std error"] = results.HC3_se
        results_df["Stars"] = results_df["Pvalues"].apply(lambda x: "***" if x < 0.01 else ("**" if x < 0.05 else ("*" if x < 0.1 else "")))
        results_df["Final"] = results_df[0].apply(lambda x: format_float(x, digits)) + results_df["Stars"] + " [" + results_df["std error"].apply(lambda x: format_float(x, digits)) + "]"
        
        f_val = format_float(results.fvalue, 3)
        p = results.f_pvalue
        if p < 0.01:
            f_val += "***"
        elif p < 0.05:
            f_val += "**"
        elif p < 0.1:
            f_val += "*"

        mean_y = format_float(y.mean(), digits)
        df_[endog] = results_df["Final"].tolist() + [mean_y, len(sub_df), format_float(results.rsquared, 3), f_val, results.aic, results.bic]
    
    return df_
```

```python
sub_df = reg_df[["Min Share Excl B",
              "Min Share Excl B_2",
              "Min Share Excl B_i",
              "Min Share Excl B_2_i",
                "Black Share", 
                "Black Share_2",
                "Rural", 
                "Midwest", 
                "Northeast", 
                "South", 
                "PC Inc", 
                "Gini", 
                "UR",
                "%BachOrMore"] + ["Loans per Emp Bus_1",
                                   "Loans per Emp Bus_2",
                                   "Loans per Emp Bus_3",
                                   "Loans per Emp Bus", 
                                   "Rural_i",
                                   "Total Pop"]].dropna()
def weighted_average(df, values_col, weights_col):
    # Element-wise multiplication, ignoring NaNs
    weighted_values = df[values_col] * df[weights_col]
    
    # Sum of weighted values and weights, ignoring NaNs
    sum_weighted_values = weighted_values.sum(skipna=True)
    sum_weights = df[weights_col].sum(skipna=True)
    
    # Calculate weighted average, returning NaN if not computable
    if sum_weights == 0:
        return np.nan
    else:
        return sum_weighted_values / sum_weights

print(weighted_average(sub_df[sub_df["Rural_i"] == 0], "Min Share Excl B", "Total Pop"))
print(weighted_average(sub_df[sub_df["Rural_i"] == 1], "Min Share Excl B", "Total Pop"))
print(weighted_average(sub_df, "Black Share", "Total Pop"))
```
