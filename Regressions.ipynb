```python
reg_df = stats_df[["Min Share", "Black Share", "Rural", 
                   "Midwest", "Northeast", "South", 
                   "PC Inc", "Gini", "UR", "Rent", 
                   "Loans per Emp Bus_1", "Loans per Emp Bus_2",
                   "Loans per Emp Bus_3", 'Avg Loan Amt per Emp_1', 
                   'Avg Loan Amt per Emp_2','Avg Loan Amt per Emp_3',
                   'Avg Loan Amt per Emp_2 Adj',
                   'Avg Loan Amt per Emp_3 Adj']].dropna()
reg_df
```

```python
from scipy import stats
alpha = 0.05  # significance level
for col in reg_df.columns:
    stat, p = stats.shapiro(reg_df[col])
    if p < alpha:
        print(f"{col} is not normally distributed (p={p:.4f})")
    else:
        print(f"{col} is normally distributed (p={p:.4f})")
```

```python
for col in reg_df.columns:
    stat, p = stats.shapiro(np.log(reg_df[col]))
    if p < alpha:
        print(f"{col} is not normally distributed (p={p:.4f})")
    else:
        print(f"{col} is normally distributed (p={p:.4f})")
```
