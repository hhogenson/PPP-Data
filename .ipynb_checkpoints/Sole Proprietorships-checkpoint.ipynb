# Investigating Sole Proprietorships

```python
import pandas as pd
import numpy as np
from datetime import datetime
import csv
import os 
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col
from scipy import stats
```

## Aggregate Data

```python
# Initialize the master DataFrame
master_df = pd.read_csv("ZIPdf.csv")
master_df.rename(columns={"Unnamed: 0" : "ZIP"}, inplace=True)
master_df["ZIP"] = master_df["ZIP"].astype(str).str.zfill(5)
master_df.set_index("ZIP", inplace=True)
master_df
```

```python
# List of file names
base = "C:\\Users\\hthog\\Desktop\\PPP Loan Data\\"
file_df = pd.read_csv(base + "PPP Files.csv")
file_df["Completed"] = 0
file_df
```

```python
def assign_value_based_on_date(date):
    if date < pd.Timestamp('2020-04-17'):
        return '_1'
    elif date < pd.Timestamp('2020-08-09'):
        return '_2'
    else:
        return '_3'
traunches = ["_1", "_2", "_3"]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
master_df["Number of Loans_1"] = 0
master_df["Total Loans Amount_1"] = 0

master_df["Number of Loans_2"] = 0
master_df["Total Loans Amount_2"] = 0

master_df["Number of Loans_3"] = 0
master_df["Total Loans Amount_3"] = 0
```

### With Sole Proprietorships

```python
sole_df = master_df.copy()
```

```python
for file in file_df[file_df["Completed"] == 0]["File"]:
    file_name = base + file
    # Check if file exists
    if not os.path.exists(file_name):
        print(f"File {file_name} does not exist. Skipping...")
        continue

    print(f"Processing {file}...")
    
    # Read file row by row since cannot fit in DF
    chunk_size = 640000
    for chunk in pd.read_csv(file_name, chunksize=chunk_size, usecols=['NonProfit', 
                                                                        'BusinessType', 
                                                                        'NAICSCode', 
                                                                        'BorrowerZip', 
                                                                        'DateApproved', 
                                                                        'CurrentApprovalAmount']):
        # Filter chunk to only needed observations
        #chunk = chunk[chunk["NonProfit"] != "Y"]
        chunk = chunk[~chunk["BusinessType"].isin(["Independent Contractors",
                                                   "Self-Employed Individuals"])]
        chunk = chunk[~chunk["NAICSCode"].astype(str).str.startswith(("482",
                                                                      "491",
                                                                      "525110",
                                                                      "525120",
                                                                      "525190",
                                                                      "525920",
                                                                      "541120",
                                                                      "814",
                                                                      "92"))]
        chunk.dropna(subset=["BorrowerZip"], inplace=True)

        # Edit chunk to include necessary info
        chunk["DateApproved"] = pd.to_datetime(chunk["DateApproved"])
        chunk["Traunch"] = chunk["DateApproved"].apply(assign_value_based_on_date)
        chunk["BorrowerZip"] = chunk["BorrowerZip"].astype(str).str[:5]
            
        # Aggregate Data
        grouped = chunk.groupby(['BorrowerZip', 'Traunch']).agg({
                'Traunch': 'size'})
        
        # Edit chunk for amount
        chunk = chunk[~chunk["NAICSCode"].astype(str).str.startswith(("111", "112"))]
        
        # Aggregate for amount
        amount_grouped = chunk.groupby(['BorrowerZip', 'Traunch']).agg({
                'CurrentApprovalAmount': 'sum'})

        # Update master_df
        for (zip_, traunch), row in grouped.iterrows():
            if zip_ in sole_df.index:
                sole_df.loc[zip_, f"Number of Loans{traunch}"] += row['Traunch']  # Update the number of loans
            else:
                continue

        for (zip_, traunch), row in amount_grouped.iterrows():
            if zip_ in sole_df.index:
                sole_df.loc[zip_, f"Total Loans Amount{traunch}"] += row['CurrentApprovalAmount']
            else:
                continue
                        
    #file_df.loc[file_df["File"] == file, "Completed"] = 1
    
    print(f"Updated master DataFrame with data from {file}.")
```

```python
sole_df["Total Loans"] = sole_df["Number of Loans_1"].add(sole_df["Number of Loans_2"], fill_value = 0).add(sole_df["Number of Loans_3"], fill_value = 0)
sole_df["Total Loan Amount"] = sole_df["Total Loans Amount_1"].add(sole_df["Total Loans Amount_2"], fill_value = 0).add(sole_df["Total Loans Amount_3"], fill_value = 0)
```

```python
sole_df[["Number of Loans_1",
           "Total Loans Amount_1",
           "Number of Loans_2",
           "Total Loans Amount_2",
           "Number of Loans_3",
           "Total Loans Amount_3",
           "Total Loans",
           "Total Loan Amount"]] = sole_df[["Number of Loans_1",
           "Total Loans Amount_1",
           "Number of Loans_2",
           "Total Loans Amount_2",
           "Number of Loans_3",
           "Total Loans Amount_3",
           "Total Loans",
           "Total Loan Amount"]].fillna(0)
```

```python
sole_df["Loans per Emp Bus"] = sole_df["Total Loans"] / sole_df["Total Emp Bus"]
sole_df["Avg Loan Amt per Emp"] = sole_df["Total Loan Amount"] / sole_df["CBP Emp"]
sole_df["Loans per Emp Bus_1"] = sole_df["Number of Loans_1"] / sole_df["Total Emp Bus"]
sole_df["Avg Loan Amt per Emp_1"] = sole_df["Total Loans Amount_1"] / sole_df["CBP Emp"]
sole_df["Loans per Emp Bus_2"] = sole_df["Number of Loans_2"] / sole_df["Total Emp Bus"]
sole_df["Avg Loan Amt per Emp_2"] = sole_df["Total Loans Amount_2"] / sole_df["CBP Emp"]
sole_df["Loans per Emp Bus_3"] = sole_df["Number of Loans_3"] / sole_df["Total Emp Bus"]
sole_df["Avg Loan Amt per Emp_3"] = sole_df["Total Loans Amount_3"] / sole_df["CBP Emp"]
sole_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
def plot_multi_binned_means(dfs, x_col, y_cols, weight_col, y_limits=None):
    # Initialize the figure
    n = len(y_cols)
    fig, axes = plt.subplots(2, n, figsize=(n * 6, 12))
    
    colors = ['blue', 'red']  # different colors for different dataframes
    
    for df_idx, df in enumerate(dfs):
        # Drop NaN values for relevant columns
        df = df.dropna(subset=[x_col] + y_cols + [weight_col])
        
        for i, y_col in enumerate(y_cols):
            for j, measure in enumerate(['Unweighted', 'Weighted']):
                ax = axes[j, i]
                title  = f'Mean {y_col}' if df_idx == 0 else '' # Set title only once
                
                # Create bins for the x column
                bin_label = f'{x_col} Bin'
                df[bin_label] = pd.cut(df[x_col], bins=20)
                
                if measure == 'Weighted':
                    # Calculate weighted mean for each bin
                    df['Weighted'] = df[y_col] * df[weight_col]
                    grouped = df.groupby(bin_label).apply(lambda x: x['Weighted'].sum() / x[weight_col].sum()).reset_index(name=f'Weighted Mean {y_col}')
                    if df_idx == 0: title += f" Weighted by {weight_col}"
                else:
                    # Calculate unweighted mean for each bin
                    grouped = df.groupby(bin_label)[y_col].mean().reset_index()
                
                # Using ax.bar to create bar charts
                bars = ax.bar(grouped[bin_label].astype(str), grouped[grouped.columns[-1]], color=colors[df_idx], alpha=0.3, label=f'DataFrame {df_idx+1}: {x_col} ({measure})')
                
                # Perform quadratic regression
                x_data = np.arange(len(grouped[bin_label]))
                y_data = grouped[grouped.columns[-1]].values
                coeffs = np.polyfit(x_data, y_data, 2)
                poly_eq = np.poly1d(coeffs)
                x_poly = np.linspace(min(x_data), max(x_data), 100)
                y_poly = poly_eq(x_poly)
                ax.plot(x_poly, y_poly, '-', color=colors[df_idx], label=f'DataFrame {df_idx+1}: Quadratic Fit')
                
                # Set the labels and title only for the first data frame
                if df_idx == 0:
                    ax.set_xticks(x_data)
                    ax.set_xticklabels(grouped[bin_label].astype(str), rotation=45)
                    title += f' by {x_col}'
                    ax.set_title(title)
                
                    if y_limits:
                        # Setting different y-limits for unweighted and weighted rows
                        ax.set_ylim(y_limits[j])
                
                ax.legend()
    
    plt.tight_layout()
    plt.show()
```

## Regressions

```python
region_mappings = {
    '9': 'West',
    '8': 'West',
    '59': 'West',
    '7': 'South',
    '885': 'South',
    '3': 'South',
    '40': 'South',
    '41': 'South',
    '42': 'South',
    '2': 'South',
    '197': 'South',
    '198': 'South',
    '199': 'South',
    '569': 'South',
    '5': 'Midwest',
    '6': 'Midwest',
    '4': 'Midwest',
    '0': 'Northeast',
    '1': 'Northeast',
}

# Define a function to map ZIP code to region based on the first few digits
def map_zip_to_region(zip_code):
    for i in range(len(zip_code), 0, -1):
        prefix = zip_code[:i]
        if prefix in region_mappings:
            return region_mappings[prefix]
    return None  # Return None if no match is found

# Apply the function to each ZIP code in the DataFrame index and create a new column
sole_df['Census Region'] = sole_df.index.map(map_zip_to_region)
dummies = pd.get_dummies(sole_df['Census Region'])
sole_df = pd.concat([sole_df, dummies], axis=1)
```

```python
def table(df, endogs, exogs, digits):
    df_ = pd.DataFrame(index = ["const"] + exogs + ["Mean","N","R2","F"], columns = endogs)
    for endog in endogs:
        sub_df = df[exogs + [endog]].dropna()
        X = sub_df[exogs]
        X = sm.add_constant(X)
        y = sub_df[endog]
        model = sm.OLS(y, X)
        results = model.fit(cov_type = 'HC3')
        results_df = pd.DataFrame(results.params) 
        results_df["Pvalues"] = results.pvalues
        results_df["std error"] = results.HC3_se
        results_df["Stars"] = results_df["Pvalues"].apply(lambda x: "***" if x < 0.01 else ("**" if x < 0.05 else ("*" if x < 0.1 else ""))) 
        results_df["Final"] = results_df[0].round(digits).astype(str) + results_df["Stars"] + " [" + results_df["std error"].round(digits).astype(str) + "]"
        f_val = str(round(results.fvalue,3))
        p = results.f_pvalue
        if p < 0.01:
            f_val += "***"
        elif p < 0.05:
            f_val += "**"
        elif p < 0.1:
            f_val += "*"
        df_[endog] = results_df["Final"].tolist() + [sub_df[endog].mean(), len(sub_df), round(results.rsquared,3), f_val]
    return df_
```

```python
table(sole_df, ["Loans per Emp Bus_1",
                "Loans per Emp Bus_2",
                "Loans per Emp Bus_3",
                "Loans per Emp Bus"], ["Min Share Excl B", 
                                        "Black Share", 
                                        "Rural", 
                                        "Midwest", 
                                        "Northeast", 
                                        "South", 
                                        "PC Inc", 
                                        "Gini", 
                                        "UR",
                                        "%BachOrMore"], 5)
```
