```python
import pandas as pd
import os
```

```python
df = pd.read_csv("Farms 1.csv")
df
```

```python
df["Zip Code"] = df["Zip Code"].astype(str).str.zfill(5)
df['Value'] = pd.to_numeric(df['Value'].str.replace(',', ''), errors='coerce')
grouped_df = df.groupby('Zip Code')['Value'].sum().reset_index()
df.rename(columns={'Value': 'Num Farms'}, inplace=True)
grouped_df.set_index('Zip Code', inplace=True)
grouped_df
```
